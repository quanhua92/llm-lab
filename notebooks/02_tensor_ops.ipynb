```python
import torch
```

```python
x = torch.tensor([[1, 2, 3], [4, 5, 6]]) # shape [2, 3]
```

```python
print(f"x: {x} \n{x.shape}")
```

```python
print("reshape: change layout without changing data")
print(f"reshaped: {x.reshape(3,2)}")
```

```python
print("transpose swaps dims 0 and 1")
print(f"transposed: {x.transpose(0,1)}")
```

```python
batched = x.unsqueeze(0)
squeezed = batched.squeeze(0)
print(f"x: \n{x},  \nshape={x.shape}")
print(f"\nunsqueeze -> batched: add dim\n{batched}, \nshape = {batched.shape}")
print(f"\nsqueeze -> remove dim: \n{squeezed}, \nshape = {squeezed.shape}")
```

```python
print(f"x: {x}")
print(f"row1: {x[1]}")
print(f"col2: {x[:, 2]}")
print(f"sub: {x[0:2, 1:3]}")
print(f"mask: {x[x > 2]} -> bool masking flatten to 1D")
```

```python
a = torch.tensor([[1, 2], [3, 4]])
b = torch.tensor([[5, 6], [7, 8]])

print("a=", a)
print("b=", b)
print(f"cat_dim0 = {torch.cat((a, b), dim=0)}")
print(f"cat_dim1 = {torch.cat((a, b), dim=1)}")
print(f"stack_dim0 = {torch.stack((a, b), dim=0)}")
print(f"stack_dim1 = {torch.stack((a, b), dim=1)}")
```

```python
row = torch.tensor([1, 2, 3])
col = torch.tensor([[10], [20]])

print(f"row: {row}\ncol: {col}")

print(f"row + col: {row + col}")
```

```python
W = torch.randn(3, 4)
v = torch.randn(1, 3)

product = torch.matmul(v, W)

print(f"\nW: {W}\n{W.shape}")
print(f"v: {v}\n{v.shape}")
print(f"torch.matmul(v, W): {product}")

v = v[0]
product = torch.matmul(v, W)

print(f"\nW: {W}\n{W.shape}")
print(f"v: {v}\n{v.shape}")
print(f"torch.matmul(v, W): {product}")
```
