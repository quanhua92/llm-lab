```python
import torch

# --- Basic autograd: track operations on a tensor ---
x = torch.tensor(3.0, requires_grad=True)
y = x ** 2                        # y = x^2, dy/dx = 2x = 6
y.backward()                      # compute gradients
print(f"x={x.item()}  y={y.item()}  dy/dx={x.grad.item()}")  # grad=6
```

```python
# --- Gradients accumulate! Must zero between steps ---
w = torch.tensor(2.0, requires_grad=True)
for i in range(3):
    loss = (w - 5) ** 2
    loss.backward()
    print(f"Step {i+1}: grad={w.grad.item()}")  # accumulates: -6, -12, -18
    w.grad.zero_()                 # COMMENT THIS OUT to see accumulation bug
```

```python
# --- no_grad: stop tracking for inference ---
model_w = torch.tensor([1.0, 2.0, 3.0], requires_grad=True)
with torch.no_grad():
    pred = model_w * 2             # no grad_fn — not tracked
print(f"Tracked? {pred.requires_grad}")  # False
```

```python
# --- The full training step (mini version) ---
data = torch.tensor([1.0, 2.0, 3.0, 4.0])
target = torch.tensor([2.0, 4.0, 6.0, 8.0])
w_train = torch.tensor(0.5, requires_grad=True)
b_train = torch.tensor(0.0, requires_grad=True)
lr = 0.01

for epoch in range(1000):
    pred = w_train * data + b_train
    loss = ((pred - target) ** 2).mean()
    loss.backward()               # compute gradients
    with torch.no_grad():
        w_train -= lr * w_train.grad   # gradient descent step
        b_train -= lr * b_train.grad
    w_train.grad.zero_()
    b_train.grad.zero_()

print(f"Learned: w={w_train.item():.3f}  b={b_train.item():.3f}")  # w≈2.0  b≈0.0
```
