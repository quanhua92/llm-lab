```python
import torch
import numpy as np
```

```python
from_list = torch.tensor([1, 2, 3])

print(f"From list:  {from_list}, {from_list.dtype}")
```

```python
np_arr = np.array([[1.0, 2.0], [3.0, 4.0]])
from_numpy = torch.from_numpy(np_arr)

print(f"From numpy: \n{from_numpy}, \nshape={from_numpy.shape}, dtype={from_numpy.dtype}")
```

```python
print(f"zeros: {torch.zeros(2, 3)}, \nones: {torch.ones(2,3)}, \nrand: {torch.rand(2,3)}")
```

```python
print(f"arange: {torch.arange(0, 10, step=1)}")
```

```python
f32 = torch.tensor([1.0, 2.0], dtype=torch.float32)
bf16 = torch.tensor([1.0, 2.0], dtype=torch.bfloat16)

print(f"f32 bytes/element_size: {f32.element_size()}, bfloat16: {bf16.element_size()}")
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
on_dev = torch.rand(3, 3, device=device)

print(f"Device: {on_dev.device}")
```

```python
x = torch.rand(4, 3, 2)
print(f"shape={x.shape} dtype={x.dtype}, device={x.device}, numel={x.numel()}")
```
